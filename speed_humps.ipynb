**Introduction**

In this project, we were trying to figure out which neighborhoods are the best for driving. To do this, we looked at different things that can affect driving. One important thing we considered was the number of speed bumps in each neighborhood. We think that having fewer speed bumps makes driving more comfortable because it means less stopping and starting. So, we're counting how many speed bumps there are in each neighborhood. This helps us see which neighborhoods have more traffic-calming measures in place. We looked at a bunch of other stuff too, but we decided to focus on speed bumps because they directly impact how fast and smooth driving feels.

**Description of metrics**

In this part of the project, our metric is to determine the best neighborhood based on the number of speed humps present. We consider neighborhoods with fewer speed humps to be better, while those with more speed humps are considered less desirable.

Datasets: To conduct this analysis, we utilized the dataset available at https://data.wprdc.org/dataset/city-of-pittsburgh-speed-humps, which contains information on all speed humps in the city of Pittsburgh up to the year 2021.

```python
import pandas as pd
%matplotlib inline
```

```python
data = pd.read_csv("speed_humps.csv")
data.head(10)
```

```python
neighborhood_counts = data.groupby("neighborhood")['id'].count()
print(neighborhood_counts)
```

```python
neighborhood_counts.plot.bar()
```
