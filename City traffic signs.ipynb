**Introduction:** To determine which neighborhoods in Pittsburgh were easier to drive on we analized the number of street signs on each street. 

**Description of metric:** We used a dataset found on Western Pennsylvania Regional Data Center titled "City Traffic Signs" (https://data.wprdc.org/dataset/city-traffic-signs). Roads containing less stop signs were deemed as 'better' to drive on due to less frequent stops.


```python
import pandas as pd


f = pd.read_csv('City traffic signs.csv')
##neighborhood, street, description

c = f[['street', 'description', 'neighborhood']]

stop = c[f['description'] == 'Stop']

nb = stop.value_counts('neighborhood')
nb
```

**Thought process markdown.**

I read in the data and focused only on columns "street", "description", and "neighborhood". (description meaning type of sign.)
Because our focus is number of stop signs, I removed any other kind of sign listed in the dataset and counted how many times a stop sign appears in each neighborhood.
Then I plotted a bar graph of each neighborhood and the frequency of stop signs.

```python
nb.plot.bar(figsize=(20,12))
```

```python
grp = stop.groupby('neighborhood').count()
grp.rename(columns= { 'description' : 'number of stop signs in neighborhood'}, inplace = True)
b = grp.sort_values(by=['number of stop signs in neighborhood'], ascending=True).head(5)

b

```

**Thought Process Markdown.**

Here I searched for the top 5 neighborhoods with the least amount of stop signs. In the table above you can see which 5 neighborhoods have the least amount of stop signs. The "street" column represents the nunber of streets in each neighborhood that contains stop signs and the "number of stop signs in neighborhood" column represents number of stop signs in each neighborhood.

**Metric Conclusion.**

Considering this metric, Mount Oliver Borough is the best neighborhood to drive in due to it only having 1 stop sign.
